```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

# Definición de la clase Pelicula para almacenar los datos de cada película
class Pelicula:
    def __init__(self, title, worldwide_earnings, domestic_earnings,
                 domestic_percent_earnings, foreign_earnings,
                 foreign_percent_earnings, year):
        self.title = title
        self.worldwide_earnings = worldwide_earnings
        self.domestic_earnings = domestic_earnings
        self.domestic_percent_earnings = domestic_percent_earnings
        self.foreign_earnings = foreign_earnings
        self.foreign_percent_earnings = foreign_percent_earnings
        self.year = year

    def __str__(self):
        return f"{self.title} ({self.year})"
    def mostrar_atributos(self):
        print(f"Title: {self.title}")
        print(f"Worldwide Earnings: {self.worldwide_earnings}")
        print(f"Domestic Earnings: {self.domestic_earnings}")
        print(f"Domestic Percent Earnings: {self.domestic_percent_earnings}")
        print(f"Foreign Earnings: {self.foreign_earnings}")
        print(f"Foreign Percent Earnings: {self.foreign_percent_earnings}")
        print(f"Year: {self.year}\n")

# Definición de la clase Nodo AVL
class NodoAVL:
    def __init__(self, key, data):
        self.key = key
        self.data = data
        self.left = None
        self.right = None
        self.height = 1

# Definición de la clase Árbol AVL
class AVLTree:
    def __init__(self):
        self.root = None

    # Función para obtener la altura de un nodo
    def get_height(self, node):
        if not node:
            return 0
        return node.height

    # Función para obtener el factor de balanceo de un nodo
    def get_balance(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)

    # Rotación a la derecha
    def rotate_right(self, y):
        x = y.left
        T2 = x.right
        x.right = y
        y.left = T2
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))
        x.height = 1 + max(self.get_height(x.left), self.get_height(x.right))
        return x

    # Rotación a la izquierda
    def rotate_left(self, x):
        y = x.right
        T2 = y.left
        y.left = x
        x.right = T2
        x.height = 1 + max(self.get_height(x.left), self.get_height(x.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))
        return y

    # Función para insertar un nodo en el árbol AVL
    def insert(self, node, key, data):
        # Paso 1: Realizar la inserción normal de BST
        if not node:
            return NodoAVL(key, data)
        elif key < node.key:
            node.left = self.insert(node.left, key, data)
        elif key > node.key:
            node.right = self.insert(node.right, key, data)
        else:
            # Duplicados no se permiten
            return node

        # Paso 2: Actualizar la altura de este nodo ancestro
        node.height = 1 + max(self.get_height(node.left), self.get_height(node.right))

        # Paso 3: Obtener el factor de balanceo de este nodo ancestro para comprobar si se desbalanceó
        balance = self.get_balance(node)

        # Si el nodo está desbalanceado, entonces hay 4 casos

        # Caso Izquierda Izquierda
        if balance > 1 and key < node.left.key:
            return self.rotate_right(node)

        # Caso Derecha Derecha
        if balance < -1 and key > node.right.key:
            return self.rotate_left(node)

        # Caso Izquierda Derecha
        if balance > 1 and key > node.left.key:
            node.left = self.rotate_left(node.left)
            return self.rotate_right(node)

        # Caso Derecha Izquierda
        if balance < -1 and key < node.right.key:
            node.right = self.rotate_right(node.right)
            return self.rotate_left(node)

        return node

    # Función para buscar un nodo en el árbol AVL
    def search(self, node, key):
        if not node or node.key == key:
            return node
        elif key < node.key:
            return self.search(node.left, key)
        else:
            return self.search(node.right, key)
    def delete(self, node, key):
    # Paso 1: Realizar la eliminación normal de BST
     if not node:
        return node

    # Si el valor a eliminar es menor que el valor de la raíz, ir a la izquierda
     if key < node.key:
        node.left = self.delete(node.left, key)
    # Si el valor a eliminar es mayor que el valor de la raíz, ir a la derecha
     elif key > node.key:
        node.right = self.delete(node.right, key)
    # Si el valor es igual a la raíz, este es el nodo a eliminar
     else:
        # Nodo con solo un hijo o sin hijos
         if not node.left:
            return node.right
         elif not node.right:
            return node.left

        # Nodo con dos hijos: obtener el sucesor inorden (más pequeño en el subárbol derecho)
         temp = self.get_min_value_node(node.right)
         node.key = temp.key
         node.data = temp.data
        # Eliminar el sucesor inorden
         node.right = self.delete(node.right, temp.key)

    # Si el árbol tiene un solo nodo, regresar el nodo
     if not node:
        return node

    # Paso 2: Actualizar la altura del nodo actual
     node.height = 1 + max(self.get_height(node.left), self.get_height(node.right))

    # Paso 3: Obtener el factor de balanceo del nodo actual
     balance = self.get_balance(node)

    # Si el nodo está desbalanceado, entonces hay 4 casos

    # Caso Izquierda Izquierda
     if balance > 1 and self.get_balance(node.left) >= 0:
        return self.rotate_right(node)

    # Caso Derecha Derecha
     if balance < -1 and self.get_balance(node.right) <= 0:
        return self.rotate_left(node)

    # Caso Izquierda Derecha
     if balance > 1 and self.get_balance(node.left) < 0:
        node.left = self.rotate_left(node.left)
        return self.rotate_right(node)

    # Caso Derecha Izquierda
     if balance < -1 and self.get_balance(node.right) > 0:
        node.right = self.rotate_right(node.right)
        return self.rotate_left(node)

     return node


    def buscar_por_titulo(self, title):
        def _buscar_nodo(node, key):
            if not node:
                return None
            if node.key == key:
                return node
            elif key < node.key:
                return _buscar_nodo(node.left, key)
            else:
                return _buscar_nodo(node.right, key)

        resultado = _buscar_nodo(self.root, title)

        if resultado:
            print(f"Nodo encontrado: {resultado.data.title}")
            resultado.data.mostrar_atributos()
        else:
            print(f"La película con el título '{title}' no se encontró en el árbol.")


    def recorrido_por_niveles(self):
      if not self.root:
        print("El árbol está vacío.")
        return

    # Utilizamos una cola para realizar el recorrido por niveles
      cola = deque([self.root])

      while cola:
        nodo_actual = cola.popleft()  # Sacamos el nodo de la cola
        print(nodo_actual.data.title)  # Mostramos el título de la película

        # Agregamos los hijos del nodo actual a la cola
        if nodo_actual.left:
            cola.append(nodo_actual.left)
        if nodo_actual.right:
            cola.append(nodo_actual.right)


    def buscar_por_criterios(self, year, min_foreign_earnings):
        resultados = []

        def _buscar_nodos(node):
            if node is not None:
                # Verificar si el nodo cumple con todos los criterios
                if (node.data.year == year and
                    node.data.domestic_percent_earnings < node.data.foreign_percent_earnings and
                    node.data.foreign_earnings >= min_foreign_earnings):
                    resultados.append(node.data)

                # Llamar recursivamente para los subárboles izquierdo y derecho
                _buscar_nodos(node.left)
                _buscar_nodos(node.right)

        _buscar_nodos(self.root)

        # Mostrar resultados
        if resultados:
            print(f"Películas encontradas con los criterios:")
            for pelicula in resultados:
                pelicula.mostrar_atributos()
        else:
            print("No se encontraron películas que cumplan con los criterios especificados.")

    def encontrar_padre(self, root, key):
     def _encontrar_padre(node, parent_key):
        if node is None:
            return None
        if node.key == parent_key:
            return node
        if key < node.key:
            return _encontrar_padre(node.left, node.key)
        else:
            return _encontrar_padre(node.right, node.key)

     parent_node = _encontrar_padre(root, key)
     if parent_node:
        return parent_node.parent
     else:
        return None
     def encontrar_abuelo(self, root, key):
      def _encontrar_nodo(node, key):
        if node is None:
            return None
        if node.key == key:
            return node
        elif key < node.key:
            return _encontrar_nodo(node.left, key)
        else:
            return _encontrar_nodo(node.right, key)

     def _encontrar_padre(node, key):
        if node is None:
            return None
        if node.key == key:
            return node.parent
        elif key < node.key:
            return _encontrar_padre(node.left, key)
        else:
            return _encontrar_padre(node.right, key)

    # Buscar el nodo que contiene la clave
     nodo = _encontrar_nodo(root, key)

     if nodo is None or nodo.parent is None:
        return None  # El nodo o su padre no existe, por lo tanto, no hay abuelo

    # Buscar el padre del nodo
     padre = nodo.parent

     if padre is None:
        return None  # El nodo no tiene padre, por lo tanto, no hay abuelo

    # Buscar el abuelo del nodo
     abuelo = padre.parent

     return abuelo
    def encontrar_tio(self, root, key):
     def _encontrar_nodo(node, key):
        if node is None:
            return None
        if node.key == key:
            return node
        elif key < node.key:
            return _encontrar_nodo(node.left, key)
        else:
            return _encontrar_nodo(node.right, key)

     def _encontrar_padre(node, key):
        if node is None:
            return None
        if node.left and node.left.key == key:
            return node
        if node.right and node.right.key == key:
            return node
        elif key < node.key:
            return _encontrar_padre(node.left, key)
        else:
            return _encontrar_padre(node.right, key)

     def _encontrar_abuelo(node, key):
        padre = _encontrar_padre(node, key)
        if padre:
            return _encontrar_padre(node, padre.key)
        return None

       # Buscar el nodo que contiene la clave
     nodo = _encontrar_nodo(root, key)

     if nodo is None:
          return None  # Nodo no encontrado

    # Buscar el padre del nodo
     padre = _encontrar_padre(root, key)

     if padre is None:
        return None  # El nodo no tiene padre, no puede tener tío

    # Buscar el abuelo del nodo
     abuelo = _encontrar_abuelo(root, key)

     if abuelo is None:
        return None  # El nodo no tiene abuelo, no puede tener tío

    # Buscar el tío del nodo
     if padre == abuelo.left:
        tio = abuelo.right
     else:
        tio = abuelo.left

     return tio







# Función para leer y buscar una película en el archivo CSV
def buscar_pelicula_y_insertar(avl_tree, nombre_pelicula, df):
    pelicula = df[df['Title'].str.lower() == nombre_pelicula.lower()]

    if not pelicula.empty:
        # Extraer los atributos necesarios y crear un objeto Pelicula
        title = pelicula.iloc[0]['Title']
        worldwide_earnings = pelicula.iloc[0]['Worldwide Earnings']
        domestic_earnings = pelicula.iloc[0]['Domestic Earnings']
        domestic_percent_earnings = pelicula.iloc[0]['Domestic Percent Earnings']
        foreign_earnings = pelicula.iloc[0]['Foreign Earnings']
        foreign_percent_earnings = pelicula.iloc[0]['Foreign Percent Earnings']
        year = pelicula.iloc[0]['Year']

        pelicula_obj = Pelicula(
            title, worldwide_earnings, domestic_earnings,
            domestic_percent_earnings, foreign_earnings,
            foreign_percent_earnings, year
        )

        # Verificar si ya está en el árbol
        if not avl_tree.search(avl_tree.root, title):
            avl_tree.root = avl_tree.insert(avl_tree.root, title, pelicula_obj)
            print(f"Película '{pelicula_obj}' insertada en el árbol AVL.")
        else:
            print(f"La película '{pelicula_obj}' ya está en el árbol AVL.")
    else:
        print("La película no se encontró en el archivo CSV.")

# Función para graficar el árbol AVL usando NetworkX y Matplotlib
def graficar_arbol(avl_tree):
    G = nx.DiGraph()  # Grafo dirigido para representar el árbol AVL

    # Función recursiva para agregar nodos y aristas al grafo
    def agregar_nodos_aristas(node):
        if node is not None:
            G.add_node(node.key, label=str(node.data))  # Agregar nodo
            if node.left is not None:
                G.add_edge(node.key, node.left.key)  # Agregar arista al hijo izquierdo
                agregar_nodos_aristas(node.left)    # Llamada recursiva para el hijo izquierdo
            if node.right is not None:
                G.add_edge(node.key, node.right.key)  # Agregar arista al hijo derecho
                agregar_nodos_aristas(node.right)     # Llamada recursiva para el hijo derecho

    # Construir el grafo desde la raíz
    agregar_nodos_aristas(avl_tree.root)

    # Dibujar el grafo usando layout de Graphviz para árbol
    try:
        pos = nx.nx_agraph.graphviz_layout(G, prog='dot')  # Usa el layout 'dot' para representación de árboles
    except ImportError:
        pos = nx.spring_layout(G)  # Fall back en caso de que pygraphviz no esté instalado

    labels = nx.get_node_attributes(G, 'label')
    nx.draw(G, pos, with_labels=True, labels=labels, node_size=3000, node_color="skyblue", font_size=8, font_weight="bold", arrows=True)
    plt.title("Árbol AVL de Películas")
    plt.show()

# Cargar el archivo CSV
try:
    df = pd.read_csv('dataset_movies.csv')
except FileNotFoundError:
    print("El archivo 'dataset_movies.csv' no se encontró.")
    exit()

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

# Definición de la clase Pelicula para almacenar los datos de cada película
class Pelicula:
    def __init__(self, title, worldwide_earnings, domestic_earnings,
                 domestic_percent_earnings, foreign_earnings,
                 foreign_percent_earnings, year):
        self.title = title
        self.worldwide_earnings = worldwide_earnings
        self.domestic_earnings = domestic_earnings
        self.domestic_percent_earnings = domestic_percent_earnings
        self.foreign_earnings = foreign_earnings
        self.foreign_percent_earnings = foreign_percent_earnings
        self.year = year

    def __str__(self):
        return f"{self.title} ({self.year})"
    def mostrar_atributos(self):
        print(f"Title: {self.title}")
        print(f"Worldwide Earnings: {self.worldwide_earnings}")
        print(f"Domestic Earnings: {self.domestic_earnings}")
        print(f"Domestic Percent Earnings: {self.domestic_percent_earnings}")
        print(f"Foreign Earnings: {self.foreign_earnings}")
        print(f"Foreign Percent Earnings: {self.foreign_percent_earnings}")
        print(f"Year: {self.year}\n")

# Definición de la clase Nodo AVL
class NodoAVL:
    def __init__(self, key, data):
        self.key = key
        self.data = data
        self.left = None
        self.right = None
        self.height = 1

# Definición de la clase Árbol AVL
class AVLTree:
    def __init__(self):
        self.root = None

    # Función para obtener la altura de un nodo
    def get_height(self, node):
        if not node:
            return 0
        return node.height

    # Función para obtener el factor de balanceo de un nodo
    def get_balance(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)

    # Rotación a la derecha
    def rotate_right(self, y):
        x = y.left
        T2 = x.right
        x.right = y
        y.left = T2
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))
        x.height = 1 + max(self.get_height(x.left), self.get_height(x.right))
        return x

    # Rotación a la izquierda
    def rotate_left(self, x):
        y = x.right
        T2 = y.left
        y.left = x
        x.right = T2
        x.height = 1 + max(self.get_height(x.left), self.get_height(x.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))
        return y

    # Función para insertar un nodo en el árbol AVL
    def insert(self, node, key, data):
        # Paso 1: Realizar la inserción normal de BST
        if not node:
            return NodoAVL(key, data)
        elif key < node.key:
            node.left = self.insert(node.left, key, data)
        elif key > node.key:
            node.right = self.insert(node.right, key, data)
        else:
            # Duplicados no se permiten
            return node

        # Paso 2: Actualizar la altura de este nodo ancestro
        node.height = 1 + max(self.get_height(node.left), self.get_height(node.right))

        # Paso 3: Obtener el factor de balanceo de este nodo ancestro para comprobar si se desbalanceó
        balance = self.get_balance(node)

        # Si el nodo está desbalanceado, entonces hay 4 casos

        # Caso Izquierda Izquierda
        if balance > 1 and key < node.left.key:
            return self.rotate_right(node)

        # Caso Derecha Derecha
        if balance < -1 and key > node.right.key:
            return self.rotate_left(node)

        # Caso Izquierda Derecha
        if balance > 1 and key > node.left.key:
            node.left = self.rotate_left(node.left)
            return self.rotate_right(node)

        # Caso Derecha Izquierda
        if balance < -1 and key < node.right.key:
            node.right = self.rotate_right(node.right)
            return self.rotate_left(node)

        return node

    # Función para buscar un nodo en el árbol AVL
    def search(self, node, key):
        if not node or node.key == key:
            return node
        elif key < node.key:
            return self.search(node.left, key)
        else:
            return self.search(node.right, key)
    def delete(self, node, key):
    # Paso 1: Realizar la eliminación normal de BST
     if not node:
        return node


    # Si el valor a eliminar es menor que el valor de la raíz, ir a la izquierda
     if key < node.key:
        node.left = self.delete(node.left, key)
    # Si el valor a eliminar es mayor que el valor de la raíz, ir a la derecha
     elif key > node.key:
        node.right = self.delete(node.right, key)
    # Si el valor es igual a la raíz, este es el nodo a eliminar
     else:
        # Nodo con solo un hijo o sin hijos
         if not node.left:
            return node.right
         elif not node.right:
            return node.left

        # Nodo con dos hijos: obtener el sucesor inorden (más pequeño en el subárbol derecho)
         temp = self.get_min_value_node(node.right)
         node.key = temp.key
         node.data = temp.data
        # Eliminar el sucesor inorden
         node.right = self.delete(node.right, temp.key)

    # Si el árbol tiene un solo nodo, regresar el nodo
     if not node:
        return node

    # Paso 2: Actualizar la altura del nodo actual
     node.height = 1 + max(self.get_height(node.left), self.get_height(node.right))

    # Paso 3: Obtener el factor de balanceo del nodo actual
     balance = self.get_balance(node)

    # Si el nodo está desbalanceado, entonces hay 4 casos

    # Caso Izquierda Izquierda
     if balance > 1 and self.get_balance(node.left) >= 0:
        return self.rotate_right(node)

    # Caso Derecha Derecha
     if balance < -1 and self.get_balance(node.right) <= 0:
        return self.rotate_left(node)

    # Caso Izquierda Derecha
     if balance > 1 and self.get_balance(node.left) < 0:
        node.left = self.rotate_left(node.left)
        return self.rotate_right(node)

    # Caso Derecha Izquierda
     if balance < -1 and self.get_balance(node.right) > 0:
        node.right = self.rotate_right(node.right)
        return self.rotate_left(node)

     return node


    def buscar_por_titulo(self, title):
        def _buscar_nodo(node, key):
            if not node:
                return None
            if node.key == key:
                return node
            elif key < node.key:
                return _buscar_nodo(node.left, key)
            else:
                return _buscar_nodo(node.right, key)

        resultado = _buscar_nodo(self.root, title)

        if resultado:
            print(f"Nodo encontrado: {resultado.data.title}")
            resultado.data.mostrar_atributos()
        else:
            print(f"La película con el título '{title}' no se encontró en el árbol.")


    def recorrido_por_niveles(self):
      if not self.root:
        print("El árbol está vacío.")
        return

    # Utilizamos una cola para realizar el recorrido por niveles
      cola = deque([self.root])

      while cola:
        nodo_actual = cola.popleft()  # Sacamos el nodo de la cola
        print(nodo_actual.data.title)  # Mostramos el título de la película

        # Agregamos los hijos del nodo actual a la cola
        if nodo_actual.left:
            cola.append(nodo_actual.left)
        if nodo_actual.right:
            cola.append(nodo_actual.right)


    def buscar_por_criterios(self, year, min_foreign_earnings):
        resultados = []

        def _buscar_nodos(node):
            if node is not None:
                # Verificar si el nodo cumple con todos los criterios
                if (node.data.year == year and
                    node.data.domestic_percent_earnings < node.data.foreign_percent_earnings and
                    node.data.foreign_earnings >= min_foreign_earnings):
                    resultados.append(node.data)

                # Llamar recursivamente para los subárboles izquierdo y derecho
                _buscar_nodos(node.left)
                _buscar_nodos(node.right)

        _buscar_nodos(self.root)

        # Mostrar resultados
        if resultados:
            print(f"Películas encontradas con los criterios:")
            for pelicula in resultados:
                pelicula.mostrar_atributos()
        else:
            print("No se encontraron películas que cumplan con los criterios especificados.")



# Función para leer y buscar una película en el archivo CSV
def buscar_pelicula_y_insertar(avl_tree, nombre_pelicula, df):
    pelicula = df[df['Title'].str.lower() == nombre_pelicula.lower()]

    if not pelicula.empty:
        # Extraer los atributos necesarios y crear un objeto Pelicula
        title = pelicula.iloc[0]['Title']
        worldwide_earnings = pelicula.iloc[0]['Worldwide Earnings']
        domestic_earnings = pelicula.iloc[0]['Domestic Earnings']
        domestic_percent_earnings = pelicula.iloc[0]['Domestic Percent Earnings']
        foreign_earnings = pelicula.iloc[0]['Foreign Earnings']
        foreign_percent_earnings = pelicula.iloc[0]['Foreign Percent Earnings']
        year = pelicula.iloc[0]['Year']

        pelicula_obj = Pelicula(
            title, worldwide_earnings, domestic_earnings,
            domestic_percent_earnings, foreign_earnings,
            foreign_percent_earnings, year
        )

        # Verificar si ya está en el árbol
        if not avl_tree.search(avl_tree.root, title):
            avl_tree.root = avl_tree.insert(avl_tree.root, title, pelicula_obj)
            print(f"Película '{pelicula_obj}' insertada en el árbol AVL.")
        else:
            print(f"La película '{pelicula_obj}' ya está en el árbol AVL.")
    else:
        print("La película no se encontró en el archivo CSV.")

# Función para graficar el árbol AVL usando NetworkX y Matplotlib
def graficar_arbol(avl_tree):
    G = nx.DiGraph()  # Grafo dirigido para representar el árbol AVL

    # Función recursiva para agregar nodos y aristas al grafo
    def agregar_nodos_aristas(node):
        if node is not None:
            G.add_node(node.key, label=str(node.data))  # Agregar nodo
            if node.left is not None:
                G.add_edge(node.key, node.left.key)  # Agregar arista al hijo izquierdo
                agregar_nodos_aristas(node.left)    # Llamada recursiva para el hijo izquierdo
            if node.right is not None:
                G.add_edge(node.key, node.right.key)  # Agregar arista al hijo derecho
                agregar_nodos_aristas(node.right)     # Llamada recursiva para el hijo derecho

    # Construir el grafo desde la raíz
    agregar_nodos_aristas(avl_tree.root)

    # Dibujar el grafo usando layout de Graphviz para árbol
    try:
        pos = nx.nx_agraph.graphviz_layout(G, prog='dot')  # Usa el layout 'dot' para representación de árboles
    except ImportError:
        pos = nx.spring_layout(G)  # Fall back en caso de que pygraphviz no esté instalado

    labels = nx.get_node_attributes(G, 'label')
    nx.draw(G, pos, with_labels=True, labels=labels, node_size=3000, node_color="skyblue", font_size=8, font_weight="bold", arrows=True)
    plt.title("Árbol AVL de Películas")
    plt.show()

# Cargar el archivo CSV
try:
    df = pd.read_csv('dataset_movies.csv')
except FileNotFoundError:
    print("El archivo 'dataset_movies.csv' no se encontró.")
    exit()



def mostrar_menu():
    print("MENÚ DE OPCIONES")
    print("1. Insertar una película")
    print("2. Eliminar una película")
    print("3. Buscar una película por título")
    print("4. Buscar películas con filtros")
    print("5. Salir")

def mostrar_menu2():
    print("MENÚ DE OPCIONES")
    print("1. Seguir insertando")
    print("2. Salir")

def menu_principal(avl_tree, df):
    while True:
        mostrar_menu()
        try:
            opcion = int(input("Seleccione una opción: "))

            if opcion == 1:
                while True:
                    nombre_pelicula = input("Ingresa el nombre de la película que deseas agregar: ")
                    # Asume que se necesita una función para agregar películas con más detalles.
                    # Por ejemplo: avl_tree.insertar_pelicula(nombre_pelicula, otros_detalles)
                    # Graficar árbol (asume que `graficar_arbol` es una función definida previamente)
                    graficar_arbol(avl_tree)

                    mostrar_menu2()
                    opcion1 = int(input("Seleccione una opción: "))
                    if opcion1 == 2:
                        break

            elif opcion == 2:
                titulo_a_eliminar = input("Ingresa el título de la película que deseas eliminar: ")
                avl_tree.root = avl_tree.delete(avl_tree.root, titulo_a_eliminar)
                avl_tree.graficar_arbol(avl_tree)
            elif opcion == 3:
                avl_tree.recorrido_por_niveles()


            elif opcion == 4:
                nombre_pelicula = input("Ingresa el nombre de la película que deseas buscar: ")
                avl_tree.buscar_por_titulo(nombre_pelicula)
                resultado = avl_tree.search(avl_tree.root, nombre_pelicula)
                if resultado:

                    print("¿Desea conocer más información de este nodo en el árbol?")
                    print("1. Sí")
                    print("2. No")
                    opcion3 = int(input("Seleccione una opción: "))
                    if opcion3 == 1:
                        # Se asume que `resultado` contiene la clave del nodo.
                        clave_buscada = resultado.key

                        balance = avl_tree.get_balance(resultado)
                        altura = avl_tree.get_height(resultado)
                        tio = avl_tree.encontrar_tio(avl_tree.root,resultado.data)
                        padre = avl_tree.encontrar_padre(avl_tree.root, resultado.data)
                        abuelo = avl_tree.encontrar_abuelo(avl_tree.root, resultado.data)

                        print(f"Balance del nodo con clave {clave_buscada}: {balance}")
                        print(f"Altura del nodo con clave {clave_buscada}: {altura}")

                        if tio:
                            print(f"El tío del nodo con clave {clave_buscada} es: {tio.key}")
                        else:
                            print(f"No se encontró el nodo con clave {clave_buscada} o no tiene tío.")

                        if padre:
                            print(f"El padre del nodo con clave {clave_buscada} es: {padre.key}")
                        else:
                            print(f"No se encontró el nodo con clave {clave_buscada} o no tiene padre.")

                        if abuelo:
                            print(f"El abuelo del nodo con clave {clave_buscada} es: {abuelo.key}")
                        else:
                            print(f"No se encontró el nodo con clave {clave_buscada} o no tiene abuelo.")
                else:
                    print("No se encontró la película con el título dado.")

            elif opcion == 5:
                year = int(input("Ingresa el año de la película que deseas buscar: "))
                domestic_percent = float(input("Ingresa el porcentaje de ingresos nacionales (domestic_percent): "))


                avl_tree.buscar_por_criterios(year, domestic_percent)

            elif opcion == 6:
                print("Gracias por usar nuestra base de datos de películas.")
                break

            else:
                print("Opción no válida. Intente de nuevo.")

        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número.")


# Llama a la función de menú principal con el árbol AVL y el DataFrame de películas
menu_principal(avl_tree, df)


```
